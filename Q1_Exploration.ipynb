```python
import pandas as pd

# Load dataset directly from GitHub (Raw URL)
url = "https://raw.githubusercontent.com/amr0ym/TM351-Data-Management-TMA/refs/heads/main/Airline%20Dataset%20Updated%20-%20v2.csv"
df = pd.read_csv(url)
```

```python
# Explore the structure and contents of the dataset

display(df.head())

print("Dataset Shape:")
print(df.shape)

print("Column Names:")
print(df.columns.tolist())

df.info()
display(df.describe(include="all"))
```

```python
# Assess data quality: completeness and consistency

print("Missing Values:")
print(df.isnull().sum())

print("Duplicate Rows:")
print(df.duplicated().sum())
```

```python
# Identify potential data inconsistencies in categorical variables

for col in df.select_dtypes(include='object').columns:
    print(f"\nUnique values in {col}:")
    print(df[col].unique())
```
